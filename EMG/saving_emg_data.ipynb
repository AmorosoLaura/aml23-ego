```python
import h5py

def stampa_dati(gruppo, indentazione=''):
    """
    Funzione ricorsiva per stampare in modo strutturato i gruppi e i dataset all'interno di un gruppo HDF5.
    """
    # Stampa il nome del gruppo
    print(indentazione + gruppo.name)

    # Se il gruppo ha attributi, stampali
    for attributo, valore in gruppo.attrs.items():
        print(indentazione + f"Attributo: {attributo}, Valore: {valore}")

    # Se il gruppo contiene altri gruppi, chiama la funzione ricorsivamente
    for nome_sotto_gruppo, sotto_gruppo in gruppo.items():
        if isinstance(sotto_gruppo, h5py.Group):
            stampa_dati(sotto_gruppo, indentazione + '  ')

        # Se il sotto-gruppo contiene un dataset, stampa il nome del dataset
        elif isinstance(sotto_gruppo, h5py.Dataset):
            print(indentazione + f"Dataset: {nome_sotto_gruppo}")


def leggi_dati_da_gruppo(gruppo):
   
    for nome_membro in gruppo:
        membro = gruppo[nome_membro]
        print(nome_membro)
        # Verifica se il membro è un gruppo
        if isinstance(membro, h5py.Group) :
            print(f"Esplorazione del gruppo '{nome_membro}':")
            # Chiamata ricorsiva per esplorare i membri del gruppo
            leggi_dati_da_gruppo(membro)
        elif isinstance(membro, h5py.Dataset):
            # Se è un dataset, leggi i primi 10 elementi
            data = membro[:10]
            print(f"Dati del dataset '{nome_membro}':")
            print(data)

# Apri il file HDF5 in modalità lettura ('r')
with h5py.File('C:/Users/Laura/Downloads/2022-06-13_22-35-11_streamLog_actionNet-wearables_S02.hdf5', 'r') as file:
    stampa_dati(file)
```

```python
import os
import numpy as np

# Percorso della cartella di cui vogliamo ottenere i nomi dei file
cartella = './EMG_data/'

# Ottieni i nomi dei file nella cartella
nomi_file = os.listdir(cartella)
videos_name=[]
# formato esempio emg-data-S05_2-left-test.pkl
for file in nomi_file:
    video=file.split("-")[2]
    train=file.split("-")[4].split(".")[0]=='train'
    if train:
        videos_name.append(video)

split="train"
print(videos_name)   
for file in nomi_file:
    is_right_split=file.split("-")[4].split(".")[0]==split
    if file.split("-")[2] in videos_name and is_right_split :
        with open(cartella+file, 'rb') as f_pickle:
            dati=pickle.load(f_pickle)
            df=pd.DataFrame(dati)

            print(df.to_string(index=False, justify="right"))
```
